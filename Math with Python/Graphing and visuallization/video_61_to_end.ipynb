```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[1 ,2] ,[3 ,4]]
plt.xticks([0 ,1])
plt.yticks([0 ,1])
plt.imshow(A)
```

```python
A = np.zeros((10 ,14))
shapeA = np.shape(A)
for i in range(0 ,shapeA[0]):
    for j in range(0 ,shapeA[1]):
        A[i ,j] = 3*i - 4*j
        
plt.plot([0 ,3] ,[9 ,2] ,'r' ,linewidth=4)
# plt.imshow(A)
for i in range(0 ,shapeA[0]):
    for j in range(0 ,shapeA[1]):
        plt.text(j ,i,int(A[i ,j]) ,horizontalalignment='center') #order is swapped because there is different order in matrix and plots
plt.set_cmap = 'twilight'
plt.imshow(A)
plt.show()
```

```python
#c[i,j] = -1^(i+j)
n = 8
C = np.zeros((n ,n))
for i in range(0 ,np.shape(C)[0]):
    for j in range(0 ,np.shape(C)[1]):
        C[i ,j] = (-1)**(i+j)
plt.imshow(C ,cmap='binary')
plt.tick_params(labelleft=False ,labelbottom=False)
plt.show()
```

# video 63 : drawing patches with polygon

```python
from matplotlib.patches import Polygon
```

```python
Y = np.array([ [1 ,1] ,[2 ,3] ,[4 ,1] ])
p = Polygon(Y ,facecolor = 'r' ,alpha = .5)

fig ,ax = plt.subplots()
ax.add_patch(p)
ax.set_ylim([0 ,6])
ax.set_xlim([0 ,6])
plt.show()
```

```python
dist = 2
square_coordinates = np.array([
    [-(dist/4) ,(dist/2 - (dist**2))],   #3*(dist/4))-(2*dist),
    [dist/4 ,(dist/2 - (dist**2))],
    [dist/4 ,-(dist**2)],
    [-(dist/4) ,-(dist**2)]
])
x = np.linspace(-dist ,dist ,100)
y = -(x**2)
curve_coords = np.vstack((x ,y)).T
square = Polygon(square_coordinates ,facecolor = 'k' ,alpha = .9)
curve = Polygon(curve_coords ,facecolor = 'b' ,alpha = .2)

fig ,ax = plt.subplots()

ax.add_patch(curve)
ax.add_patch(square)
plt.plot(x ,y ,'k')
plt.plot([x[0] ,x[-1]] ,[y[0] ,y[-1]] ,'k')
# plt.axis('square')
plt.show()
```
