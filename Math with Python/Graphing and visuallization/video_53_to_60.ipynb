# video 56 : graphing lines - slope-intercept form

```python
#y = mx + b
x = [-5 ,5]
m = 2
b = 5
y = [1 ,2]
for i in range(0 ,len(x)):
    y[i] = m*x[i] + b

plt.plot(x ,y)
plt.xlim(x)
plt.ylim(x)
axis = plt.gca()
plt.plot(axis.get_xlim() ,[0 ,0] ,'k--')
plt.plot([0 ,0] ,axis.get_xlim() ,'k--')
plt.axis('square')
plt.grid()
plt.show()
```

# video 57 : rational function

```python
x = range(-3 ,4)
y = np.zeros(len(x))
for i in range(0 ,len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x ,y ,'s-')
plt.show()
```

```python
# resolution
x = np.linspace(-3 ,4 ,10)
y = 2 + np.sqrt(abs(x))

plt.plot(x ,y )
plt.show()
```

```python
powers = range(-1 ,4)
x = np.linspace(-4 ,4 ,60)
y = np.zeros(len(x))
for p in powers:
    #p is the power
    y = x**(p)
    plt.plot(x ,y ,label='$y=x^{%s}$' %p ,linewidth=4, )

plt.legend()
plt.xlim([-4 ,4])
plt.ylim([-4 ,4])
plt.show()
```

# video 58 : plotting with sympy

```python
import sympy as sym
from sympy.abc import x ,y
```

```python
y = x**6
p = sym.plotting.plot(y ,2*y)
```

```python
import sympy.plotting.plot as symplot
x,a = sym.symbols('x a')
expr = a/x
p = symplot(expr.subs(a ,1),(x,-5 ,5) ,show=False)
p[0].label = '$y = %s$' %sym.latex((expr.subs(a ,1)))

p2 = symplot(expr.subs(a ,2),(x,-5 ,5) ,show=False)
p.extend(p2)
p[0].line_color = 'r'
p.legend = True
p.ylim = [-5 ,5]
p.show()
```

# video 59 : sympy exercises

```python
y = a/(x**2 - a)
p = None
p = symplot(y.subs(a ,1) ,(x ,-5 ,5) ,show=False)
p[0].label = '$%s$' %sym.latex(y.subs(a ,1))

for i in range(2 ,5):
    p.extend(symplot(y.subs(a ,i) ,(x ,-5 ,5) ,show=False))
    p[i-1].label = '$%s$' %sym.latex(y.subs(a ,i))
    p[i-1].line_color = [np.random.rand(3)]
    
p.ylim = [-10 ,10]
p.legend = True
p.show()
```
