```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

# video 103 : plotting sine ,cosine ,tangent and resolution

```python
x = np.linspace(0 ,6*np.pi ,100)
plt.plot(x ,np.sin(x) ,label = 'sin($\\theta$)')
plt.plot(x ,np.cos(x) ,label = 'cos($\\theta$)')

plt.xlabel('rad')
plt.xlabel('f(x)')
plt.legend()
plt.show()
```

```python
th = np.linspace(0 ,4*np.pi ,300)
plt.plot(th ,np.tan(th) ,label = 'tan($\\theta$)')

plt.xlabel('rad')
plt.xlabel('f(x)')
plt.ylim([-5 ,5])
plt.legend()
plt.show()
```

```python
th = np.linspace(0 ,4*np.pi ,3000)

y1 = np.sin(th + np.cos(th)**2)
y2 = np.cos(th + np.sin(th))

plt.plot(th ,y1 ,label = 'sin($\\theta$ + cos($\\theta$))')
plt.plot(th ,y2 ,label = 'cos($\\theta$ + sin($\\theta$))')
plt.xlabel('rad')
plt.xlabel('f(x)')
plt.legend()
plt.show()
```

```python
thsym = sym.pi*7/6
theta = float(thsym)

#unit circle
x = np.linspace(0 ,2*np.pi ,200)
plt.plot(np.cos(x) ,np.sin(x) ,color='gray')
plt.plot([-1.1 ,1.1] ,[0 ,0] ,'--' ,color='gray')
plt.plot([0 ,0] ,[-1.1 ,1.1] ,'--' ,color='gray')

#arc
plt.plot(np.cos(np.linspace(0 ,theta ,100)) ,np.sin(np.linspace(0 ,theta ,100)) ,'k') 
plt.plot([0 ,np.cos(theta)] ,[0 ,np.sin(theta)] ,'v--')
plt.plot(np.cos(theta) ,np.sin(theta) ,'ro')
plt.plot([0 ,0] ,[0 ,np.sin(theta)] ,'--',color='green')
plt.plot([0 ,np.cos(theta)] ,[np.sin(theta) ,np.sin(theta)] ,'--',color='red')


plt.title('$\\theta = %s$'%(sym.latex(thsym)))
plt.axis('square')
plt.grid()
plt.show()
```

# video 105 and onwards : Euler's formula and exercises

```python
k = np.pi/4
m = 2.3
j = complex(0 ,1)
eul = m*np.exp(j*k)
cis = m*np.cos(k) + m*j*np.sin(k)
print(eul)
print(cis)
eul == cis
```

```python
mag = np.abs(eul)
ang = np.angle(eul)
plt.polar([0 ,ang] ,[0 ,mag] ,'r')
```

```python
def eulerfromcossin():
    re = eval(input("Cosine part : "))
    im = eval(input("Sine part : "))
    m = np.sqrt(re**2 + im**2)
    k = np.arctan2(im ,re)
    
    plt.polar([0 ,k] ,[0 ,m] ,'r')
    plt.show()
```

```python
n = 64
m = np.random.rand(n)
#now we need to distribute it between 0 and 1 which it already is
k = np.random.rand(n)*(2*np.pi)
#need to distribute it between 0 and 360deg ie 0 and 2*np.pi

for i in range(n):
    r = np.random.rand()
    if r<.4:
        clr = [1 ,.2 ,.7]
    elif r>.4 and r<.8:
        clr = [.7 ,.2 ,1]
    else:
        clr = [0 ,0 ,1]
    plt.polar([0 ,k[i]] ,[0 ,m[i]] ,color=clr)
    plt.polar(k[i] ,m[i] ,'o' ,color=clr)

plt.axis(False)
plt.show()
```

```python
#Parametric functions

r1 = np.random.rand()
r2 = np.random.rand()
theta = np.linspace(0 ,4*np.pi ,100)
x = np.cos(r1*theta)
y = np.sin(r2*theta)

plt.plot(x ,y ,'--')
plt.title('r1 = %.2f ,r2 = %.2f' %(r1 ,r2))
plt.axis('off')
plt.show()
```
