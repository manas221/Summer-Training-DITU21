```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display ,Math
```

```python
nums = np.random.rand(1000)
plt.plot(nums ,'s')
#uniform distribution
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000)
plt.plot(nums ,'s' )
#uniform distribution
plt.show()

plt.hist(nums)
plt.show()
```

```python
# exercise : generate normal dis. with mean = 15 and SD = 4.3
desired_mean = 15
desired_sd = 4.3
nums = np.random.randn(1000)  # * sd + mean
nums = nums - nums.mean()
nums = desired_sd * (nums/nums.std())
nums = nums + desired_mean  #do this after adjusting sd
plt.subplot(121)  #1 row ,2 col ,1st subplot
plt.plot(nums)

plt.subplot(122)
plt.hist(nums)

plt.show()

print("Mean :: %s \nSD :: %s" %(np.mean(nums) ,np.std(nums)))
```

#  video 99 : plot random phase angles

```python
length = 10
lims = np.random.rand(100)
lims = lims*(2*length) - length
n = 100
for i in range(n):
    coords = np.random.choice(lims ,2)
    plt.plot([0 ,coords[0]] ,[0 ,coords[1]])
    plt.plot(coords[0] ,coords[1] ,'o')
    
plt.axis('off')
plt.show()
```

```python
angs = np.random.randn(100)
for i in angs:
    plt.plot([0 ,np.cos(i)] ,[0 ,np.sin(i)])
    plt.plot(np.cos(i) ,np.sin(i) ,'o')

plt.axis('off')
plt.axis('square')
plt.show()
```

# video 101 : exercise for v100

```python
def angle_convertplot():
    angle =  eval(input("What angle to convert ? "))
    form = str(input("What unit you input in? rad or deg?"))
    inrad = 0.0
    if (form == 'deg'):
        inrad = np.pi * angle / 180
    elif (form == 'rad'):
        inrad = angle
    else:
        print("Enter correct input")
    plt.plot([0 ,1.0] ,[0 ,0] ,'k' ,linewidth=3)
    plt.plot(1.0 ,0 ,'o')
    plt.plot([0 ,np.cos(inrad)] ,[0 ,np.sin(inrad)] ,'k' ,linewidth=3)
    plt.plot(np.cos(inrad) ,np.sin(inrad) ,'o')

    plt.grid()
    plt.axis('square')
    plt.axis([-1.0 ,1.0 ,-1.0 ,1.0])
    plt.show()
```

```python
angle_convertplot()
```

# video 102 : Pythagorean theoram

```python
a = 3
b = 4
c = np.sqrt(a**2 + b**2)

plt.plot([0 ,a] ,[0 ,0])
plt.plot([0 ,0] ,[0 ,b])
plt.plot([0 ,a] ,[b ,0])
plt.show()
```
