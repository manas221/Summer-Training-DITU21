```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
h = 2
k = -2
x = np.linspace(h-10 ,h+10 ,100)
y = a*((x-h)**2) + k
d = k - (1/(4*a))
plt.plot(x ,y ,label='parabola')
plt.plot(h ,k ,'ro' ,label='vertex')
plt.plot(h ,k+(1/(4*a)) ,'bo' ,label='focus')
plt.plot([x[0] ,x[-1]] ,[d ,d] ,'k--')


plt.legend()
plt.show()
```

# Video 92 : graphing circles

```python
h = 0
k = 0
r = 1
axlim = r + np.max((abs(h) ,abs(k)))
x = np.linspace(-2*axlim ,2*axlim ,100)
X ,Y = np.meshgrid(x ,x)
Fxy = (x-h)**2 + (Y-k)**2 - r**2
plt.contour(X ,Y ,Fxy,0) #traces the values which has same elvation if you assume Fxy to be elevation of x-y matrix
plt.plot(h ,k ,'go')
plt.plot([-2*axlim ,2*axlim] ,[0 ,0] ,'k--')
plt.plot([0 ,0] ,[-2*axlim ,2*axlim] ,'k--')

plt.gca().set_aspect('equal')
plt.show()
```

```python
hs = [-1.5 ,1.5]
k = 0
radii = np.linspace(.5 ,3 ,15)
axlim = np.max(r) + np.max((abs(np.max(hs)) ,abs(k)))
x = np.linspace(-5 ,5 ,100)
X ,Y = np.meshgrid(x ,x)
for h in hs:
    for r in radii:
        Fxy = (x-h)**2 + (Y-k)**2 - r**2
        plt.contour(X ,Y ,Fxy,0 ,colors=[r/10 ,r/4 ,r/5])

plt.plot(hs ,[0 ,0] ,'k')
plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

# graphing ellipses using mesh grids

```python
h = 1
k = 2 
a = 2
b = 3

axlim = np.max((a ,b)) + np.max((abs(h) ,abs(k)))
x = np.linspace(-axlim ,axlim ,100)
X,Y = np.meshgrid(x ,x)
ellipse = ((X-h)/a)**2 + ((Y-k)/b)**2 - 1

plt.contour(X ,Y ,ellipse ,0)
plt.gca().set_aspect('equal')
plt.show()
```

```python
n = 16
a = np.linspace(-4 ,4 ,n)
b = 4
h = 0
k = np.linspace(-4 ,4 ,n)

x = np.linspace(-8 ,8 ,100)
X,Y = np.meshgrid(x ,x)
for i in range(0 ,n):
    F  = ((X-h)/a[i])**2 + ((Y-k[i])/b)**2 - 1
    plt.contour(X,Y ,F ,0 ,colors=[0 ,i/n ,i/n])

plt.show()
```

# video 94 : Hyperbolas 

```python
a = 1
b = .5
h = 1
k = 2
axlim = np.max((a ,b)) + np.max((abs(h) ,abs(k)))
x = np.linspace(-axlim ,axlim)
X ,Y = np.meshgrid(x ,x)
hyperbola = ((X-h)/a)**2 - ((Y-k)/b)**2 - 1
plt.contour(X ,Y ,hyperbola ,0)
plt.gca().set_aspect('equal')
plt.show()
```
