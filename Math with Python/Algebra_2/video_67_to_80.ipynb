# Video 71 : The quadratic equation

```python
a = 3
b = 7
c = 5
#imaginary with np
quadeqP = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)
quadeqN = (-b - np.sqrt(b**2 - 4*a*c)) / (2*a)
print(quadeqP ,quadeqN)
```

```python
a = 3
b = 7
c = 5
#imaginary with scipy
quadeqP = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
quadeqN = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)
print(quadeqP ,quadeqN)
```

```python
def quadeq(a ,b ,c):
    sol = np.zeros(2 ,dtype=complex)   #initialize zeros
    sol[0] = (-b + np.lib.scimath.sqrt(b**2 - 4*a*c)) / (2*a)
    sol[1] = (-b - np.lib.scimath.sqrt(b**2 - 4*a*c)) / (2*a)
    return sol
```

```python
print(quadeq(1 ,2 ,1))
```

# video 76 : revisiting qudratic equation

For more details, see [quadratic_matrix](#quadratic_matrix)

```python
aex = 1
bex = range(-5 ,6)
cex = range(-2 ,11)
IMG = np.zeros([len(bex) ,len(cex)] ,dtype=complex)
print('Real and Complex roots of a quadratic equation')
for i in range(0 ,len(bex)):
    for j in range(0 ,len(cex)):
        IMG[i ,j] = quadeq(aex ,bex[i] ,cex[j])[0]
        
plt.subplot(1 ,3 ,1)
plt.imshow(IMG.real ,extent = [cex[0] ,cex[-1] ,bex[0] ,bex[-1]])
plt.title("real")
plt.axis('on')

plt.subplot(1 ,3 ,2)
plt.imshow(IMG.imag ,extent = [cex[0] ,cex[-1] ,bex[0] ,bex[-1]])
plt.title("imaginary")
plt.axis('off')

plt.subplot(1 ,3 ,3)
plt.imshow(np.absolute(IMG) ,extent = [cex[0] ,cex[-1] ,bex[0] ,bex[-1]])
plt.title("magnitude")
plt.colorbar()
plt.axis('off')

plt.show()
```

# Exercise ,vid 80 : A family of gaussians

```python
#basically we have to make an image where each pixel relates to a gaussian curve
#each row has different value of h
#the more close to zero ,the brighter the pixel
```

```python
N = 100
x = np.linspace(-3 ,3 ,1001)
G = np.zeros((N ,len(x)))
h = np.zeros(N)
for i in range(N):
    num = -4*np.log(2)*(x**2)
    h[i] = (i+1)/N
    G[i ,:] = np.exp(num/(h[i]**2))
plt.pcolormesh(x,h ,G)
plt.show()
```
