# AP and GP : Video 84

```python
a_initial = 2
g_initial = 1
d = 3
r = 2
maxn = 10
AP = a_initial + d*np.arange(0 ,maxn)
GP = g_initial*(r**np.arange(0 ,maxn))
plt.plot(np.arange(0 ,maxn) ,AP ,'rs' ,label = 'Arithmetic seq')
plt.plot(np.arange(0 ,maxn) ,GP ,'bs' ,label = 'Geometric seq')
plt.grid()
plt.legend()
plt.show()
```

# Video 86 : maxima and minima of functions

```python
lst = [-4 ,5 ,4 ,1 ,2]
np.argmax(lst)   #returns index of largest element
```

```python
x = np.linspace(0 ,2*np.pi ,20)
fx = -(np.sqrt(x) + np.cos(x))

fmax = np.max(fx)
fmaxidx = np.argmax(fx)
plt.plot(x ,fx ,'bo-')
plt.plot(x[fmaxidx] ,fmax ,'rs')
plt.show()
```

```python
x = np.linspace(-2 ,2 ,51)
fx = x**3 + x**4 
fminidx = np.argmin(fx)
plt.plot(x ,fx ,'bs-' )
plt.plot(x[fminidx] ,fx[fminidx] ,'ko')

plt.xlim([-1.2 ,0.5])
plt.ylim([-.2 ,.1])
plt.show()
```

```python
x = np.linspace(0 ,12*np.pi ,100)
fx = -(np.sqrt(x) + np.cos(x))
peaks ,d = find_peaks(fx ,distance = 2)
plt.plot(x ,fx ,'b')
for peak in peaks:
    plt.plot(x[peak] ,fx[peak] ,'o')

plt.grid()
plt.show()
```

# video 87 : odd and even functions

```python
x = np.linspace(-5 ,5 ,80)
feven = np.cos(x)
feven_neg = np.cos(-x)
plt.plot(x ,feven)
plt.plot(x ,feven_neg ,'ro')
plt.show()

fodd = np.sin(x)
fodd_neg = np.sin(-x)
plt.plot(x ,fodd)
plt.plot(x ,fodd_neg ,'go-')
plt.show()
```

```python
# cos is even while sin is an odd function
```
