```python
import numpy as np
import matplotlib.pyplot as plt
import sympy.plotting.plot as symplot
import sympy as sym
from IPython.display import display, Math
```

# V23 : trig derivative

```python
q = sym.symbols('q')
fx = sym.cos(q)

for i in range(0 ,4):
    display(Math('\\frac{d}{dx} (%s) = %s' %(sym.latex(fx) ,sym.latex(sym.diff(fx))) ))
    fx = sym.diff(fx)
```

```python
f = sym.cos(q)
# p = symplot(f ,show=False ,label=sym.latex(f) )
clrs = 'krbg'
for i in range(0 ,4):
    if (i == 0):
        p = symplot(f ,show=False ,label= sym.latex(f) ,line_colors=(i/3 ,i/3 ,i/3) )
    else:
        p.extend( symplot(f ,show = False ,label = sym.latex(f) ,line_colors=(i/3 ,i/3 ,i/3)) )
    f = sym.diff(f)

p.legend = True
p.xlim = [-3 ,3]
p.show()
```

```python
x = sym.symbols('x')
fx = sym.cos(x + sym.sin(x) ) 
for a in range(0 ,4):
    fax = fx + a
    if (a == 0):
        p = symplot(fax ,show=False ,label= '$ %s $' %(sym.latex(fax)) ,line_color=(a/4 ,a/5 ,a/4) )
    else:
        p.extend( symplot(fax ,show = False ,label = '$ %s $' %(sym.latex(fax)) ,line_color=(a/4 ,a/5 ,a/4)) )


p.legend = True
# p.xlim = [-3 ,3]
p.show()
```

# Vid 125 : tangents

```python

xx = np.linspace(-3 ,3 ,252)
x = sym.symbols('x')
f = x**2
fxx = sym.lambdify(x ,f)
# plt.plot(xx ,fxx(xx) ,'k--' ,linewidth=4)

#tangent at xn
xnr = np.linspace(-2 ,2 ,150)
for xn in xnr:
#     clr = [np.random.rand() ,np.random.rand() ,np.random.rand()]
    clr = [0.1 ,0.2 ,1]
    yy = float(sym.diff(f).subs(x ,xn)) * (xx - xn) + fxx(xn)
#     plt.plot(xn ,fxx(xn) ,'o' ,c=clr)
    plt.plot(xx ,yy ,c=clr)

plt.plot(xx ,fxx(xx) ,'k' ,linewidth=4)
plt.axis([-3 ,3 ,-6 ,6])
plt.show()
```
